```python
# 1) Write a Python program to find those numbers which are divisible by 7 and multiple of 5,
#  between 1500 and 2700 (both included).

print("The numbers that are divisible by 7 and multiple of 5, between 1500 and 2700 (both included) are :: ")
for i in range(1500, 2701):
    if(i%7 == 0 and i%5 == 0):
        print(i)
```

```python
# 3) Write a Python program that prints each item and its corresponding type from the following list. 
# Sample List : datalist = [1452, 11.23, 1+2j, True, 'w3resource', (0, -1), [5, 12], {"class":'V', "section":'A'}]

datalist = [1452, 11.23, 1+2j, True, 'w3resource', (0, -1), [5, 12], {"class":'V', "section":'A'}]

for item in datalist:
    print("Item :: ", item)
    print("Type :: ", type(item))
    print("\n")
```

```python
# 4) Write a Python program that accepts a string and calculate the number of digits and letters. 
# Sample Data : Python 3.2 Letters 6 Digits 2


# sample_data = input("Enter a String to calculate the number of digits and letters :: ")
sample_data = "Python 3.2 Letters 6 Digits 2"

num_of_digits=num_of_letters=0
for char in sample_data:
    if char.isdigit():
        num_of_digits += 1
    elif char.isalpha():
        num_of_letters += 1
    else:
        pass
    
print("Number of Letters :: ", num_of_letters)
print("Number of Digits :: ", num_of_digits)
```

```python
# 5) Write a Python program to check a triangle is equilateral, isosceles or scalene.
# Note : An equilateral triangle is a triangle in which all three sides are equal. 
#     A scalene triangle is a triangle that has three unequal sides. 
#     An isosceles triangle is a triangle with (at least) two equal sides. 
#     Expected Output: Input lengths of the triangle sides: x: 6 y: 8 z: 12 Scalene triangle
                        
    
print("Please input lengths of the sides of the triangle(x, y, z)\n")
x = input("x :: ")
y = input("y :: ")
z = input("z :: ")

if (x == y == z):
    print("\nInput lengths of the triangle sides: x: ", x, " y: ", y, " z: ", z, " \nIt is an Equilateral Triangle\n")
elif (x == y or y == z or z == x):
    print("\nInput lengths of the triangle sides: x: ", x, " y: ", y, " z: ", z, " \nIt is an Isosceles Triangle\n")
else:
    print("\nInput lengths of the triangle sides: x: ", x, " y: ", y, " z: ", z, " \nIt is an Scalene Triangle\n")
```
