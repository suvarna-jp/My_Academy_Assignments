# RE Assignment 

In this assignment, you'll be working with messy medical data and using regex to extract relevant infromation from the data. 

Each line of the `dates.txt` file corresponds to a medical note. Each note has a date that needs to be extracted, but each date is encoded in one of many formats.

The goal of this assignment is to correctly identify all of the different date variants encoded in this dataset and to properly normalize and sort the dates. 

Here is a list of some of the variants you might encounter in this dataset:
* 04/20/2009; 04/20/09; 4/20/09; 4/3/09
* Mar-20-2009; Mar 20, 2009; March 20, 2009;  Mar. 20, 2009; Mar 20 2009;
* 20 Mar 2009; 20 March 2009; 20 Mar. 2009; 20 March, 2009
* Mar 20th, 2009; Mar 21st, 2009; Mar 22nd, 2009
* Feb 2009; Sep 2009; Oct 2010
* 6/2008; 12/2009
* 2009; 2010

Once you have extracted these date patterns from the text, the next step is to sort them in ascending chronological order accoring to the following rules:
* Assume all dates in xx/xx/xx format are mm/dd/yy
* Assume all dates where year is encoded in only two digits are years from the 1900's (e.g. 1/5/89 is January 5th, 1989)
* If the day is missing (e.g. 9/2009), assume it is the first day of the month (e.g. September 1, 2009).
* If the month is missing (e.g. 2010), assume it is the first of January of that year (e.g. January 1, 2010).
* Watch out for potential typos as this is a raw, real-life derived dataset.

With these rules in mind, find the correct date in each note and return a regular list in chronological order of the original Series' indices.

```python
import re, datetime

### File open dates.txt
f = open("005_dates.txt", "r")

txt = f.read()

### Declare empty date list
dt_list = []
error_dt_list = []

### Method to extract & append dates
def extractDates(dates_list):

    global dt_list
    global error_dt_list
    print(len(dates_list))

    for i in dates_list:
        
#         print("i : ", i)
        str_list = str(i).split('/')

#         print("str_list : ", str_list)
        dt = int(str_list[1])
        if dt > 31:
            print("Date Exceeding 31. Considering date error, hence skipping... ", dt)
            print("Skipping...date.... : ", i)
            error_dt_list.append(i)
            continue;

        mn = int(str_list[0])
        if mn > 12 or mn < 1:
            print("Month not in range 1...12. Considering date error, hence skipping... ", mn)
            print("Skipping...date.... : ", i)
            error_dt_list.append(i)
            continue;

        yr = int(str_list[2])
        if (yr < 100):
            yr += 1900

        full_dt = datetime.date(yr, mn, dt)

        dt_list.append(full_dt)
        
    return dt_list

### Method to extract & append dates
def extractDates_pattern2(dates_list):

    global dt_list
    global error_dt_list
    print(len(dates_list))
    
    def getMonth(month):
        if(month.upper() == 'JAN' or month.upper() == 'JANUARY'):
            return 1
        elif(month.upper() == 'FEB' or month.upper() == 'FEBUARY'):
            return 2
        elif(month.upper() == 'MAR' or month.upper() == 'MARCH'):
            return 3
        elif(month.upper() == 'APR' or month.upper() == 'APRIL'):
            return 4
        elif(month.upper() == 'MAY' or month.upper() == 'MAY'):
            return 5
        elif(month.upper() == 'JUN' or month.upper() == 'JUNE'):
            return 6
        elif(month.upper() == 'JUL' or month.upper() == 'JULY'):
            return 7
        elif(month.upper() == 'AUG' or month.upper() == 'AUGUST'):
            return 8
        elif(month.upper() == 'SEP' or month.upper() == 'SEPTEMBER'):
            return 9
        elif(month.upper() == 'OCT' or month.upper() == 'OCTOBER'):
            return 10
        elif(month.upper() == 'NOV' or month.upper() == 'NOVEMBER'):
            return 11
        elif(month.upper() == 'DEC' or month.upper() == 'DECEMBER'):
            return 12
        else:
            return 0
        

    for i in dates_list:
        
#         print("i : ", i)
        str_list = str(i).split('-')

#         print("str_list : ", str_list)
        dt = int(str_list[1])
        if dt > 31:
            print("Date Exceeding 31. Considering date error, hence skipping... ", dt)
            print("Skipping...date.... : ", i)
            error_dt_list.append(i)
            continue;

        mn = getMonth(str_list[0])
        if mn > 12 or mn < 1:
            print("Month not in range 1...12. Considering date error, hence skipping... ", mn)
            print("Skipping...date.... : ", i)
            error_dt_list.append(i)
            continue;

        yr = int(str_list[2])
        if (yr < 100):
            yr += 1900

        full_dt = datetime.date(yr, mn, dt)

        dt_list.append(full_dt)
        
    return dt_list

### Specify Date pattern for search 4/3/09
p = re.compile('\d/\d/\d\d') #4/3/09

### Find occurances of date pattern
s_list=(p.findall(txt))
extractDates(s_list)
print(len(dt_list))
# dates.append(extractDates(s_list))


### Repeat for other patterns of date 4/20/09
p = re.compile('\d/\d\d/\d\d') #4/20/09
s_list=(p.findall(txt))

extractDates(s_list)
print("New List :: ",len(dt_list))

### Repeat for other patterns of date 04/20/09
p = re.compile('\d\d/\d\d/\d\d') #04/20/09
s_list=(p.findall(txt))

extractDates(s_list)
print("New List :: ",len(dt_list))

### Repeat for other patterns of date 04/20/2009
p = re.compile('\d\d/\d\d/\d\d\d\d') #04/20/2009
s_list=(p.findall(txt))

extractDates(s_list)
print("New List :: ",len(dt_list))

### Repeat for other patterns of date Mar-20-2009; Mar 20, 2009; March 20, 2009; Mar. 20, 2009; Mar 20 2009;
p = re.compile('\w\w\w-\d\d-\d\d\d\d') #Mar-20-2009
p = re.compile('\w\w\w \d\d, \d\d\d\d') #Mar 20, 2009
p = re.compile('\w+ \d\d, \d\d\d\d') #March 20, 2009

s_list=(p.findall(txt))

extractDates_pattern2(s_list)
print("New List :: Pattern 2 :: ",len(dt_list))
    
print(len(dt_list))
print("error_dt_list Len :: ", len(error_dt_list), " :: ", error_dt_list)

def displayDates(dt_list):
    for d in dt_list:
         print(d)

displayDates(dt_list)
dt_list.sort()
print("\n\nSorted List :: ")
displayDates(dt_list)
```
